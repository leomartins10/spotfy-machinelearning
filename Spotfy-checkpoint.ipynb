### Pré-Processamento dos dados de treino

```python
def artist_hitmaker(df):
    df['artists_list'] = df['artists'].str.split(';')
    artist_stats = {}
    for _, row in df.iterrows():
        artists = row['artists_list']
        is_hit = row['popularity_target']
        
        weight = 1 / len(artists)
        
        for artist in artists:
            if artist not in artist_stats:
                artist_stats[artist] = {'total_tracks': 0, 'hit_tracks': 0}
            
            artist_stats[artist]['total_tracks'] += weight
            artist_stats[artist]['hit_tracks'] += weight * is_hit

    for artist in artist_stats:
        stats = artist_stats[artist]
        if stats['total_tracks'] > 0:
            artist_stats[artist]['hitmaker'] = stats['hit_tracks'] / stats['total_tracks']
        else:
            artist_stats[artist]['hitmaker'] = 0

    df['artist_hitmaker'] = df['artists_list'].apply(
        lambda artists: sum(artist_stats[artist]['hitmaker'] for artist in artists) / len(artists)
    )


    return df


artist_hitmaker(df)
df.head(3)
```

```python
from sklearn.preprocessing import StandardScaler, PowerTransformer, QuantileTransformer, MinMaxScaler
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
from category_encoders import TargetEncoder

# Aplicar PowerTransformer (Box-Cox) nas colunas onde faz sentido
pt = PowerTransformer(method='box-cox', standardize=False)
df['duration_ms'] = pt.fit_transform(df[['duration_ms']])

# Aplicar transformação logarítmica
log_transform_cols = ['energy', 'speechiness', 'valence']
for col in log_transform_cols:
    df[col] = np.log1p(df[col])

# Aplicar transformação cúbica em 'acousticness' e 'liveness'
df['acousticness'] = np.cbrt(df['acousticness'])
df['liveness'] = np.cbrt(df['liveness'])

# Aplicar QuantileTransformer para redistribuição uniforme de 'loudness' e 'instrumentalness'
qt = QuantileTransformer(output_distribution='normal')
df['loudness'] = qt.fit_transform(df[['loudness']])
df['instrumentalness'] = qt.fit_transform(df[['instrumentalness']])

# Aplicar a transformação quadrática
square_transform_cols = ['danceability', 'loudness']
for col in square_transform_cols:
    df[col] = df[col] ** 2

cols = ['duration_ms', 'energy', 'speechiness', 'valence', 'acousticness', 'liveness', 'loudness', 'instrumentalness', 'danceability']

# Target Encoding das colunas clusterizadas usando popularity_target
target_enc = TargetEncoder()
for col in cols:
    # Agrupar a variável numérica em bins
    df[col] = pd.cut(df[col], bins=5, labels=False)

    # Aplicar Target Encoding nos bins
    df[col] = target_enc.fit_transform(df[col], df['popularity_target'])


df['artist_frequency'] = df['artists'].map(df['artists'].value_counts())

# Calcular 'hits_per_genre' e 'genre_hitmaker'
hits_per_genre = df[df['popularity_target'] == 1].groupby('track_genre').size()
total_per_genre = df.groupby('track_genre').size()
genre_hitmaker = hits_per_genre / total_per_genre
df['genre_hitmaker'] = df['track_genre'].map(genre_hitmaker)

# Frequência de artistas

# Visualizar o resultado
df.head()
```

```python
# Split the 'artists' column to handle multiple artists per track (feats)
df['artists_list'] = df['artists'].str.split(';')

# Create a new dataframe to store artist statistics
artist_stats = {}

# Iterate over each row to accumulate stats for each artist
for _, row in df.iterrows():
    artists = row['artists_list']
    is_hit = row['popularity_target']
    
    # Each artist gets 1 / number_of_artists as their contribution
    weight = 1 / len(artists)
    
    for artist in artists:
        if artist not in artist_stats:
            artist_stats[artist] = {'total_tracks': 0, 'hit_tracks': 0}
        
        # Update the total tracks and hit tracks with the weighted values
        artist_stats[artist]['total_tracks'] += weight
        artist_stats[artist]['hit_tracks'] += weight * is_hit

# Calculate the hitmaker ratio for each artist
for artist in artist_stats:
    stats = artist_stats[artist]
    if stats['total_tracks'] > 0:
        artist_stats[artist]['hitmaker'] = stats['hit_tracks'] / stats['total_tracks']
    else:
        artist_stats[artist]['hitmaker'] = 0

# Now, let's map these hitmaker values back to the original dataframe
# Each row will have the average hitmaker value of all its artists
df['artist_hitmaker'] = df['artists_list'].apply(
    lambda artists: sum(artist_stats[artist]['hitmaker'] for artist in artists) / len(artists)
)

# Display the updated dataframe to the usera
# import ace_tools as tools; tools.display_dataframe_to_user(name="Updated Music Data", dataframe=df)

df.head()
```

### Pré-processamento dos dados de teste

```python
df_test = df_test.merge(df[['artists','artist_hitmaker', 'artist_frequency']].drop_duplicates(), on='artists', how='left')
```

```python
# 1. Criar um dicionário com o 'track_genre' e seu respectivo 'genre_hitmaker'
genre_hitmaker_dict = df.set_index('track_genre')['genre_hitmaker'].to_dict()

# 2. Usar o .map() para adicionar a coluna 'genre_hitmaker' no df_test
df_test['genre_hitmaker'] = df_test['track_genre'].map(genre_hitmaker_dict)

# 3. Preencher os valores NaN com 0.5 (caso algum gênero de df_test não exista em df)
df_test['genre_hitmaker'].fillna(0.5, inplace=True)

# Exibir as primeiras linhas para verificar o resultado
# print(df_test.head())
```

```python
from sklearn.preprocessing import PowerTransformer, QuantileTransformer
from category_encoders import TargetEncoder
import numpy as np
import pandas as pd

# Suponha que df seja seu conjunto de treino e df_test seja seu conjunto de teste

# Separar as colunas que precisam de transformações
log_transform_cols = ['energy', 'speechiness', 'valence']
cubic_transform_cols = ['acousticness', 'liveness']
quantile_transform_cols = ['loudness', 'instrumentalness']
square_transform_cols = ['danceability', 'loudness']

# 1. Ajustar os transformadores com o treino e depois aplicar no teste
# PowerTransformer (Box-Cox)
pt = PowerTransformer(method='box-cox', standardize=False)

# 2. Aplicar a transformação logarítmica
for col in log_transform_cols:
    df[col] = np.log1p(df[col])
    df_test[col] = np.log1p(df_test[col])

# 3. Aplicar a transformação cúbica
for col in cubic_transform_cols:
    df[col] = np.cbrt(df[col])
    df_test[col] = np.cbrt(df_test[col])

# 4. Aplicar o QuantileTransformer
qt = QuantileTransformer(output_distribution='normal')
for col in quantile_transform_cols:
    df[col] = qt.fit_transform(df[[col]])  # Fit no treino
    df_test[col] = qt.transform(df_test[[col]])  # Transform no teste

# 5. Aplicar a transformação quadrática
for col in square_transform_cols:
    df[col] = df[col] ** 2
    df_test[col] = df_test[col] ** 2

# 6. Aplicar Target Encoding (Fit no treino, Transform no teste)
target_enc = TargetEncoder()

cols = ['duration_ms', 'energy', 'speechiness', 'valence', 'acousticness', 'liveness', 'loudness', 'instrumentalness', 'danceability']

for col in cols:
    # Bin the columns (mesma abordagem no treino e no teste)
    df[col] = pd.cut(df[col], bins=5, labels=False)
    df_test[col] = pd.cut(df_test[col], bins=5, labels=False)

    df[col] = target_enc.fit_transform(df[col], df['popularity_target'])  # Fit no treino
    df_test[col] = target_enc.transform(df_test[col])  # Apenas transform no teste
```

```python
from sklearn.metrics import  accuracy_score, confusion_matrix

y_pred = rnf.predict(X_test)

acc = accuracy_score(y_test, y_pred)

print(acc)
```
